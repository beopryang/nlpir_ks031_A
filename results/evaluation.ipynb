```python
import pandas as pd

# ê°™ì€ í´ë”(í˜„ì¬ ì‘ì—… ë””ë ‰í† ë¦¬)ì— ìˆë‹¤ê³  ê°€ì •
naive_path = "eval_top5_truth_naive_top5_absolute.csv"
onto_path  = "eval_top5_truth_onto_top5_absolute.csv"

# CSV ë¡œë“œ
df_naive = pd.read_csv(naive_path)
df_onto  = pd.read_csv(onto_path)

# ì¶œë ¥
print("=== Naive RAG í‰ê°€ ë°ì´í„° ===")
display(df_naive.head())

print("=== Ontology RAG í‰ê°€ ë°ì´í„° ===")
display(df_onto.head())

print("Naive shape:", df_naive.shape)
print("Ontology shape:", df_onto.shape)

```

```python
# ================================================
# ì˜¨í†¨ë¡œì§€/ë‚˜ì´ë¸Œ RAG í‰ê°€ ë°ì´í„° ì „ì²˜ë¦¬ ì…€
#  - ì‚¬ì‹¤ì„± ì˜¤ë¥˜ ì •ì˜: WRONG_COUNCIL / WRONG_PERSON / WRONG_PARTY
#  - ì˜¨í†¨ë¡œì§€ EMPTY_SEGMENT ì§ˆì˜ëŠ” ì•„ì˜ˆ ì œì™¸
#  - í•´ë‹¹ ì§ˆì˜ì˜ Naive ê²°ê³¼ëŠ” ì‚¬ì‹¤ì„± ì˜¤ë¥˜ + ê´€ë ¨ì„± 0ì  íŒ¨ë„í‹°
# ================================================

# 0) ê¸°ë³¸ ì„¤ì •
FACT_ERROR_TYPES = {"WRONG_COUNCIL", "WRONG_PERSON", "WRONG_PARTY"}

# -------------------------------
# 1) WRONG_* ì‚¬ì‹¤ì„± ì˜¤ë¥˜ ê°œìˆ˜ íŒŒì•… (ì›ë³¸ ê¸°ì¤€)
# -------------------------------
naive_total_before = len(df_naive)
onto_total_before  = len(df_onto)

mask_naive_wrong = df_naive["error_type"].isin(FACT_ERROR_TYPES)
mask_onto_wrong  = df_onto["error_type"].isin(FACT_ERROR_TYPES)

num_naive_wrong = int(mask_naive_wrong.sum())
num_onto_wrong  = int(mask_onto_wrong.sum())

print("=== WRONG_* ì‚¬ì‹¤ì„± ì˜¤ë¥˜ ê°œìˆ˜ (ì›ë³¸ ê¸°ì¤€) ===")
print(f"Naive  WRONG_* ë‹¨ë½ ìˆ˜: {num_naive_wrong} / {naive_total_before}")
print(f"Onto   WRONG_* ë‹¨ë½ ìˆ˜: {num_onto_wrong} / {onto_total_before}")
print("--------------------------------------------")

# -------------------------------
# 2) ì˜¨í†¨ë¡œì§€ EMPTY_SEGMENT ì§ˆì˜ íŒŒì•… ë° ì œê±°
# -------------------------------
mask_empty_onto = df_onto["error_type"] == "EMPTY_SEGMENT"
empty_qids = sorted(df_onto.loc[mask_empty_onto, "query_id"].unique())
empty_qid_set = set(empty_qids)
num_empty_rows_onto = int(mask_empty_onto.sum())

print(f"EMPTY_SEGMENT ì§ˆì˜ ê°œìˆ˜: {len(empty_qids)}")
print("EMPTY_SEGMENT query_id ëª©ë¡:", empty_qids)
print(f"EMPTY_SEGMENTì— í•´ë‹¹í•˜ëŠ” Ontology ë‹¨ë½ ìˆ˜: {num_empty_rows_onto}")
print("--------------------------------------------")

# ì˜¨í†¨ë¡œì§€ ë°ì´í„°ì—ì„œ EMPTY_SEGMENT í–‰ ì•„ì˜ˆ ì œê±°
df_onto = df_onto.loc[~mask_empty_onto].copy()
onto_total_after = len(df_onto)

# -------------------------------
# 3) fact_ok ìƒì„±
#    - Ontology: WRONG_* ë§Œ fact_ok = 0, ë‚˜ë¨¸ì§€ëŠ” 1
#    - Naive   : WRONG_* ë˜ëŠ” EMPTY_SEGMENT ì§ˆì˜ì— ì†í•˜ëŠ” í–‰ â†’ fact_ok = 0
# -------------------------------

# (1) Ontology fact_ok
df_onto["fact_ok"] = df_onto["error_type"].apply(
    lambda e: 0 if e in FACT_ERROR_TYPES else 1
)

onto_fact_error = int((df_onto["fact_ok"] == 0).sum())

print("[Ontology] fact_ok ìƒì„± í›„")
print(f"  ì‚¬ì‹¤ì„± ì˜¤ë¥˜(fact_ok=0) ë‹¨ë½ ìˆ˜: {onto_fact_error} / {onto_total_after}")
print("--------------------------------------------")

# (2) Naive fact_ok
mask_naive_empty = df_naive["query_id"].isin(empty_qid_set)

# ì´ì „ì— fact_okê°€ ìˆì—ˆë‹¤ë©´ ë°±ì—…
if "fact_ok" in df_naive.columns:
    df_naive["fact_ok_before"] = df_naive["fact_ok"]
else:
    df_naive["fact_ok_before"] = None

# ê¸°ë³¸ê°’ 1ë¡œ ì´ˆê¸°í™” í›„, WRONG_* ë˜ëŠ” EMPTY_SEGMENT ì§ˆì˜ëŠ” 0
df_naive["fact_ok"] = 1
df_naive.loc[mask_naive_wrong | mask_naive_empty, "fact_ok"] = 0

naive_fact_error = int((df_naive["fact_ok"] == 0).sum())

# ê²¹ì¹˜ëŠ” ê²½ìš°(WRONG_* ì´ë©´ì„œ EMPTY_SEGMENT ì§ˆì˜)ë„ í™•ì¸
mask_naive_both       = mask_naive_wrong & mask_naive_empty
mask_naive_wrong_only = mask_naive_wrong & ~mask_naive_empty
mask_naive_empty_only = mask_naive_empty & ~mask_naive_wrong

num_naive_both       = int(mask_naive_both.sum())
num_naive_wrong_only = int(mask_naive_wrong_only.sum())
num_naive_empty_only = int(mask_naive_empty_only.sum())

print("[Naive] fact_ok ìƒì„± í›„")
print(f"  ì‚¬ì‹¤ì„± ì˜¤ë¥˜(fact_ok=0) ë‹¨ë½ ìˆ˜: {naive_fact_error} / {naive_total_before}")
print("  â”œ WRONG_* ë§Œ í•´ë‹¹ (EMPTY_SEGMENT ì•„ë‹˜):", num_naive_wrong_only)
print("  â”œ EMPTY_SEGMENT ì§ˆì˜ë§Œ í•´ë‹¹          :", num_naive_empty_only)
print("  â”” ë‘˜ ë‹¤ í•´ë‹¹(WRONG_* âˆ§ EMPTY_SEGMENT):", num_naive_both)
print("    (ìœ„ ì„¸ ê°œì˜ í•©ì´ fact_ok=0 ì´ ê°œìˆ˜ì´ë©°, ê° í–‰ì€ í•œ ë²ˆë§Œ ì¹´ìš´íŠ¸ë¨)")
print("--------------------------------------------")

# -------------------------------
# 4) Naive ê´€ë ¨ì„± ì ìˆ˜ íŒ¨ë„í‹° ì ìš©
#    - EMPTY_SEGMENT ì§ˆì˜ì˜ ëª¨ë“  í–‰: topic_score, overall_score = 0.0
#    - ë³€í™” ì „ ì ìˆ˜ëŠ” topic_before, overall_beforeì— ë°±ì—…
# -------------------------------
df_naive["topic_before"]   = df_naive["topic_score"]
df_naive["overall_before"] = df_naive["overall_score"]

df_naive.loc[mask_naive_empty, "topic_score"]   = 0.0
df_naive.loc[mask_naive_empty, "overall_score"] = 0.0

penalty_rows = int(mask_naive_empty.sum())

print(f"[Naive] EMPTY_SEGMENT ì§ˆì˜ íŒ¨ë„í‹° ì ìš©ëœ ë‹¨ë½ ìˆ˜: {penalty_rows}")
print("--------------------------------------------")

# -------------------------------
# 5) ìµœì¢… ìš”ì•½ ì¶œë ¥
# -------------------------------
print("=== ìµœì¢… ìš”ì•½ ===")
print(f"Ontology ë‹¨ë½ ìˆ˜:  ì „ì²˜ë¦¬ ì „ {onto_total_before}  â†’ ì „ì²˜ë¦¬ í›„ {onto_total_after}")
print(f"Naive   ë‹¨ë½ ìˆ˜:  ì „ì²˜ë¦¬ ì „ {naive_total_before} â†’ ì „ì²˜ë¦¬ í›„ {naive_total_before} (í–‰ ìˆ˜ëŠ” ë™ì¼, ì ìˆ˜/flagë§Œ ë³€ê²½)")
print(f"Ontology ì‚¬ì‹¤ì„± ì˜¤ë¥˜(fact_ok=0) ë‹¨ë½ ìˆ˜: {onto_fact_error}")
print(f"Naive   ì‚¬ì‹¤ì„± ì˜¤ë¥˜(fact_ok=0) ë‹¨ë½ ìˆ˜: {naive_fact_error}")
print("--------------------------------------------")

print("[Ontology] ì‚¬ì‹¤ì„± ì˜¤ë¥˜ í–‰ ì˜ˆì‹œ:")
display(df_onto.loc[df_onto["fact_ok"] == 0, ["query_id", "rank", "error_type", "fact_ok"]].head(10))

print("[Naive] ì‚¬ì‹¤ì„± ì˜¤ë¥˜ í–‰ ì˜ˆì‹œ:")
display(df_naive.loc[df_naive["fact_ok"] == 0, ["query_id", "rank", "error_type", "fact_ok"]].head(10))

print("[Naive] EMPTY_SEGMENT ì§ˆì˜ íŒ¨ë„í‹° í–‰ ì˜ˆì‹œ:")
display(
    df_naive.loc[mask_naive_empty, [
        "query_id", "rank",
        "topic_before", "topic_score",
        "overall_before", "overall_score",
        "error_type", "fact_ok"
    ]].head(10)
)
```

# ğŸ“— **Interpretation of Results (English)**

## 1. Redefinition of factual correctness

The original LLM-based `fact_score` was not used as-is.

**Only the following are treated as factual errors (`fact_ok=0`):**

* WRONG_COUNCIL
* WRONG_PERSON
* WRONG_PARTY

This definition aligns better with the core factual checks in the legislative-minute domain.

### âœ” Therefore:

* Ontology RAG: these 85 rows were **removed** before evaluation
